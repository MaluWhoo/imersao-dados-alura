```python
# prompt: trazuda para português as categorias da coluna "senioridade" do dataframe DF

senioridade = {
    'SE': 'Senior',
    'MI': 'Pleno',
    'EN': 'Junior',
    'EX': 'Executivo'
}
df['senioridade'] = df['senioridade'].replace(senioridade)
df['senioridade'].value_counts()
```

```python
contrato = {
    'FT': 'Tempo Integral',
    'CT': 'Contrato',
    'PT': 'Tempo Parcial',
    'FL': 'Freelancer'
}
df['contrato'] = df['contrato'].replace(contrato)
df['contrato'].value_counts()
```

```python
tamanho_empresa = {
    'M': 'Médio',
    'L': 'Grande',
    'S': 'Pequeno'
}
df['tamanho_empresa'] = df['tamanho_empresa'].replace(tamanho_empresa)
df['tamanho_empresa'].value_counts()
```

```python
remoto = {
    0: 'Presencial',
    50: 'Híbrido',
    100: 'Remoto'
}
df['remoto'] = df['remoto'].replace(remoto)
df['remoto'].value_counts()
```

```python
# Criação de um Data Frame

import numpy as np
df_salarios = pd.DataFrame({
    'nome': ['Anna', 'Miau', 'Harry', 'Rony', 'Val'],
    'salario': [4000, np.nan, 5000, np.nan, 100000] # defidindo um valor nulo
})

## Calculo da Média e substitui os nulos pela média e arredonda os valores
# df_salarios - é a nossa base
# ['salario_media'] - é a coluna nova que estamos criando

# dentro da base df_salarios, vamos preencher (fillna). Com o que será preenchido: com os salarios .mean que é a média
# .round2 - arredondar para duas casas decimais
df_salarios ['salario_media'] = df_salarios ['salario'].fillna(df_salarios ['salario'].mean().round(2))

## Calculo de Mediana e substitui os nulos pela mediana

df_salarios ['salario_mediana'] = df_salarios ['salario'].fillna(df_salarios ['salario'].median())
df_salarios
```

```python
# Outro Data Frame

df_temperaturas = pd.DataFrame({
    'Dia': ['Segunda-Feira', 'Terça-Feira', 'Quarta-Feira', 'Quinta-feira', 'Sexta-Feira'], # Criando as colunas
    'Temperatura': [30, np.nan, np.nan, 28, 27]
})

#

df_temperaturas['Preenchido_ffill'] = df_temperaturas['Temperatura'].ffill() # Criando uma nova coluna
df_temperaturas
```
