```python
# Cria√ß√£o de um Data Frame

import numpy as np
df_salarios = pd.DataFrame({
    'nome': ['Anna', 'Miau', 'Harry', 'Rony', 'Val'],
    'salario': [4000, np.nan, 5000, np.nan, 100000] # defidindo um valor nulo
})

## Calculo da M√©dia e substitui os nulos pela m√©dia e arredonda os valores
# df_salarios - √© a nossa base
# ['salario_media'] - √© a coluna nova que estamos criando

# dentro da base df_salarios, vamos preencher (fillna). Com o que ser√° preenchido: com os salarios .mean que √© a m√©dia
# .round2 - arredondar para duas casas decimais
df_salarios ['salario_media'] = df_salarios ['salario'].fillna(df_salarios ['salario'].mean().round(2))

## Calculo de Mediana e substitui os nulos pela mediana

df_salarios ['salario_mediana'] = df_salarios ['salario'].fillna(df_salarios ['salario'].median())
df_salarios
```

# Desafio da Aula 3:



Cargo - Data Science - Gerar uma visualiza√ß√£o de salario m√©dio por pa√≠s usando o gr√°fico interativo.

```python
# Filtrar o DataFrame para cargos de 'Data Scientist'
df_data_scientist = df_limpo[df_limpo['cargo'] == 'Data Scientist'].copy()

# Calcular o sal√°rio m√©dio por pa√≠s
salario_por_pais = df_data_scientist.groupby('empresa')['usd'].mean().reset_index()

# Criar o gr√°fico de barras interativo
fig = px.bar(
    salario_por_pais,
    x='empresa',
    y='usd',
    title='Sal√°rio M√©dio de Data Scientists por Pa√≠s',
    labels={'empresa': 'Pa√≠s', 'usd': 'Sal√°rio M√©dio Anual (USD)'},
    color='empresa'
)

fig.show()
```

```python
import pycountry

# Fun√ß√£o converter ISO-2 para ISO-3 = Converte sigla de pa√≠ses com 2 letras para 3.
def iso_to_iso3(code):
  try:
    return pycountry.countries.get(alpha_2=code).alpha_3
  except:
    return None

# Nova coluna com a sigla ISO 3
df_limpo ['residencia_iso3'] = df_limpo['residencia'].apply(iso_to_iso3)

# Calcular M√©dia Salarial por pa√≠s (ISO-3)
df_ds = df_limpo[df_limpo['cargo'] == 'Data Scientist']
media_por_pais = df_ds.groupby('residencia_iso3')['usd'].mean().reset_index()

# Gerar o mapa
fig = px.choropleth(
    media_por_pais,
    locations='residencia_iso3',
    color='usd',
    color_continuous_scale='rdylgn',
    title='Sal√°rio M√©dio por Pa√≠s (Data Scientist)',
    labels={'usd': 'Sal√°rio M√©dio (USD)', 'residencia_iso3': 'Pa√≠s'}
)


fig.show()
```
