```python
# feita a partir da IA

renomear_colunas = {
    'ano': 'ano',
    'senioridade':  'senioridade',
    'contrato': 'contrato',
    'cargo': 'cargo',
    'salario': 'salario',
    'moeda': 'moeda',
    'usd': 'usd',
    'residencia': 'residencia',
    'remoto': 'remoto',
    'empresa': 'empresa',
    'tamanho_empresa': 'tamanho_empresa'
}

df.rename(columns=renomear_colunas, inplace=True)
df.columns
```

```python
# prompt: trazuda para português as categorias da coluna "senioridade" do dataframe DF

senioridade = {
    'SE': 'Senior',
    'MI': 'Pleno',
    'EN': 'Junior',
    'EX': 'Executivo'
}
df['senioridade'] = df['senioridade'].replace(senioridade)
df['senioridade'].value_counts()
```

```python
contrato = {
    'FT': 'Tempo Integral',
    'CT': 'Contrato',
    'PT': 'Tempo Parcial',
    'FL': 'Freelancer'
}
df['contrato'] = df['contrato'].replace(contrato)
df['contrato'].value_counts()
```

```python
tamanho_empresa = {
    'M': 'Médio',
    'L': 'Grande',
    'S': 'Pequeno'
}
df['tamanho_empresa'] = df['tamanho_empresa'].replace(tamanho_empresa)
df['tamanho_empresa'].value_counts()
```

```python
remoto = {
    0: 'Presencial',
    50: 'Híbrido',
    100: 'Remoto'
}
df['remoto'] = df['remoto'].replace(remoto)
df['remoto'].value_counts()
```
